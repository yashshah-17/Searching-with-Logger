```python
# Importing the necessary packages
import os
import pandas as pd
import csv
import re
import time


# Defining and opening a csv file to store the data
csvfile = open("Books.csv", 'wt', newline='', encoding="utf-8")
writer = csv.writer(csvfile)


# Defining the headers inside the csv file
csvrow = []
csvrow.append("Title")
csvrow.append("Author")
writer.writerow(csvrow)


# Defining and opening a csv file to store the timing data
csvfile2 = open("Timings.csv", 'wt', newline='', encoding="utf-8")
writer2 = csv.writer(csvfile2)


# Defining the headers inside the Timings.csv file
csvrow2 = []
csvrow2.append("File_Name")
csvrow2.append("Start")
csvrow2.append("End")
writer2.writerow(csvrow2)


# Reading the text files, and inserting title and author into the csv file
for i in range(1996, 2021):
    csvrow2 = []
    
    text_file = "GUTINDEX." + str(i) + ".txt"
    csvrow2.append(text_file)
    
    local_time = time.ctime(time.time())
    csvrow2.append(local_time)
    
    flag = 0
    with open(text_file, 'rt', encoding="utf-8") as readfile:
        for readline in readfile:
            csvrow = []
    
            if(readline != "TITLE and AUTHOR                                                     ETEXT NO.\n"):
                if(flag == 0 or readline == "\n" or "[" in readline or ", by " not in readline):
                    continue
                
                else:
                    valid_line = readline.split(", by ")
                    csvrow.append(valid_line[0].strip())
            
                    author_name = re.sub('  .*\n', '', valid_line[1])
                    csvrow.append(author_name.strip())
                    
                    writer.writerow(csvrow)
                    continue
            flag = 1
    
    local_time = time.ctime(time.time())
    csvrow2.append(local_time)
    
    writer2.writerow(csvrow2)
    
    print(text_file + " -> Done")                                # Indicator to know that processing of file is done
#     time.sleep(1)                                              # Introducing a delay of 5 minutes after processing each file
           

# Closing the csv file
csvfile.close()
csvfile2.close()
```
